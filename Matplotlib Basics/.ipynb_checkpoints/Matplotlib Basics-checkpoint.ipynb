```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,11)
```

```python
y = 2 * x
```

```python
plt.plot(x,y) # this will not work in a normal editor, only works in Jupyter Notebook
```

```python
# to do it in a normal editor, you must also write plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xlim(0,7) # start and end of the x values of the line
plt.ylim(0,15) # start and end of the y values of the line
plt.title('Plot')
# plt.savefig('firstplot.png') 
# I commented the above line out because every time I will run this, it will save a new image
# Uncomment if needed
# You can see the image in the folder that this file is in
```
