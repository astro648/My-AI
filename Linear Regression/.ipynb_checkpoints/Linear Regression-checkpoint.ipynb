```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
addf = pd.read_csv('advertising.csv')
```

```python
addf.head()
```

```python
addf['Total Spent'] = addf['TV'] + addf['Radio'] + addf['Newspaper']
```

```python
sns.regplot(data=addf,x='Total Spent',y='Sales') # regression plot
```

```python
X = addf['Total Spent']
y = addf['Sales']
np.polyfit(X,y,deg=1)
```
