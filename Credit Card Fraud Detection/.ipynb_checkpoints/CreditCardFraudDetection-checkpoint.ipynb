# Credit Card Fraud Detection Project

```python
import pandas as pd
import numpy as np
import keras
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
cc_csv = pd.read_csv('creditcard.csv')
```

```python
cc_csv.head()
```

```python
# Normalize the 'Amount' column using StandardScaler
cc_csv['NormalizedAmount'] = StandardScaler().fit_transform(cc_csv['Amount'].values.reshape(-1,1))

# Drop the original 'Amount' column
cc_csv = cc_csv.drop(['Amount'], axis=1)
cc_csv = cc_csv.drop(['Time'],axis=1)
```

```python
cc_csv.head()
```

```python
# Assign features (independent variables) to X
X = cc_csv.iloc[:, cc_csv.columns != 'Class']

# Assign target (dependent variable) to y
y = cc_csv.iloc[:, cc_csv.columns == 'Class']
```

```python
X.head()
```

```python
y.head()
```

```python
# X being capital and y being lowercase is a convention
# test_size = 0.3 means 30/70 split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3,random_state = 0) # Splitted the classes into these two
```

```python
X_train = np.array(X_train) #converted X_train data to numpy array
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
# Creates a 6-layer neural network
model = Sequential([
    Dense(units = 16,input_dim=29,activation = 'relu'),
    Dense(units = 24,activation = 'relu'),
    Dropout(0.5), # If any parameter is not according to our requirements, we will drop it here
    Dense(20,activation = 'relu'),
    Dense(24,activation = 'relu'),
    Dense(1,activation = 'sigmoid'), # use sigmoid instead of relu to calculate values
])
```

```python
model.summary() # We will train on these parameters
```

```python
# Training
# optimizer optimizes code and gets weight of neural network
# binary_crossentropy makes the output either a zero or a one
# accuracy is the percentage of accurate outputs
model.compile(optimizer = 'adam',loss='binary_crossentropy',metrics=['accuracy'])
# epochs means how many times it will run
# batch size means how much data it will use
model.fit(X_train,y_train,batch_size = 15,epochs = 5)
```

```python
score = model.evaluate(X_test, y_test)
```
