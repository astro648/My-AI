```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head() # the beginning parts of the data
```

```python
# figsize: change sizing of the plot
# dpi: changes resolution of the plot
plt.figure(figsize = (12,4),dpi=200)

# x, y show the values that will be used
# data is the source dataframe
# hue changes the color depending on the data provided
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education')
```

```python
plt.figure(figsize = (12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='salary') # different columns for hue cause different results
```

```python
plt.figure(figsize = (12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Set1')
# Now it uses the Set1 color palette from the Qualitative Colormaps section https://matplotlib.org/stable/tutorials/colors/colormaps.html
```

```python
sns.rugplot(x='salary',data=df, height=0.5) # Basic rug plot
```

```python
sns.set(style = 'whitegrid') # this changes the style of the plot

sns.displot(data=df,x='salary',bins=20,color='red') # this is a histogram
# bins will change the amount of bars
# if you set bins to 20, then there will be 20 bars
# color changes color (of course)
```
